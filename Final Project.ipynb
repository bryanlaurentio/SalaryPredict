```python
import pandas as pd

df = pd.read_csv('Pendapatan_train.csv')
df_test = pd.read_csv('Pendapatan_test.csv')
df
```

```python
df[['id','Umur','Kelas Pekerja','Berat Akhir','Pendidikan','Jmlh Tahun Pendidikan','Status Perkawinan','Pekerjaan','Jenis Kelamin','Keuntungan Kapital','Kerugian Capital','Jam per Minggu','Gaji']].isnull().sum()
```

```python
df['Kelas Pekerja'].unique()
```

```python
#terdapat '?' didalam kolom kelas pekerja jadi harus di clean dulu
import numpy as np
df['Kelas Pekerja'].replace('?', np.nan, inplace = True)
df_test['Kelas Pekerja'].replace('?', np.nan, inplace = True)

df['Kelas Pekerja'].unique()
```

```python
df['Pekerjaan'].unique()
```

```python
#terdapat '?' didalam kolom pekerjaan jadi harus di clean dulu
df['Pekerjaan'].replace('?', np.nan, inplace = True)
df_test['Pekerjaan'].replace('?', np.nan, inplace = True)

df['Pekerjaan'].unique()
```

```python
df.fillna(method = 'ffill', inplace = True)
df_test.fillna(method = 'ffill', inplace = True)
```

```python
df['Pendidikan'].unique()
```

```python
#dibuat dalam kategori
obj_pendidikan = {
                  '1st-4th':0,
                  '5th-6th':1,
                  '7th-8th':2,
                  '9th':3,
                  '10th':4,
                  '11th':5,
                  '12th':6,
                  'SD':7,
                  'SMA':8,
                  'Pendidikan Tinggi':9,
                  'D3':10,
                  'D4':11,
                  'Sarjana':12,
                  'Master':13,
                  'Doktor':14,
                  'Sekolah Professional':15
}

obj_gaji = {'>7jt':1, '<=7jt':0}
df['Pendidikan'].replace(obj_pendidikan, inplace = True)
df_test['Pendidikan'].replace(obj_pendidikan, inplace = True)

df['Gaji'].replace(obj_gaji, inplace = True)
```

```python
df = pd.get_dummies(df, columns=['Kelas Pekerja', 'Status Perkawinan', 'Pekerjaan', 'Jenis Kelamin'])

df_test = pd.get_dummies(df_test, columns=['Kelas Pekerja', 'Status Perkawinan', 'Pekerjaan', 'Jenis Kelamin'])
```
